Now we are trying to collect tweets with the keywords "Magic Leap" by using tweepy. In order to have an information back-up storage when analyzing the data, we store all data we scrapped from tweeter to mongoDB (credentials can be found here: https://github.com/Nancyxw/AI_Final_Project/blob/master/credentials1.txt). 

Here we only print the tweet create date, full text, user id, retweets and favorites. In this way, readers could get first impression of raw tweets data.

Data cleaning will be done in the next notebook.

```python
#Scrapping
import json
import pymongo
import tweepy

with open('consumer_key.txt', 'r') as f:
    consumer_key =  f.read()
f.closed

with open('consumer_secret.txt', 'r') as f:
    consumer_secret = f.read()
f.closed

with open('access_key.txt', 'r') as f:
    access_key = f.read()
f.closed

with open('access_secret.txt', 'r') as f:
     access_secret = f.read()
f.closed


#Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
api = tweepy.API(auth)

print("success!")

#connect to mongoDB
try:
    #use your database name, user and password here:
    #mongodb://<dbuser>:<dbpassword>@<mlab_url>/<database_name>
    with open("credentials1.txt", 'r', encoding='utf-8') as f:
        [name,password,url,dbname]=f.read().splitlines()
        conn=pymongo.MongoClient("mongodb://{}:{}@{}/{}".format(name,password,url,dbname))
        print ("Connected successfully!!!")
except pymongo.errors.ConnectionFailure as e:
    print ("Could not connect to MongoDB: %s" % e) 

db = conn['aifinalproject_1']
collection = db['Magic_Leap']

#Start scarpping - "Magic Leap"
Data = []
for tweet in tweepy.Cursor(api.search,q="Magic Leap",
                           lang="en",
                           since="2017-12-18",
                           until="2018-01-03",
                           tweet_mode="extended").items():
    tweet_data = tweet._json
    collection.insert_one(tweet_data)
    Data.append(tweet_data)
    print (tweet.created_at, tweet.full_text, tweet.user.id, "Retweets",tweet.retweet_count, "Favorites",tweet.favorite_count)

print("Finish!!!")
```

```python
#convert data to csv
import pandas as pd
df = pd.DataFrame(Data)
df.to_csv("magicleap.csv",encoding = 'utf-8')
```

#### Brief Summary: By using tweepy, we successfully collected 5018 tweets with the keyword "Magic Leap" from 20/12/2017 to 01/03/2018 and saved them into "magicleap.csv".
